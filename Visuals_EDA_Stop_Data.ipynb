### Visuals

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stop_data = pd.read_csv('../CTA_Data/stop_totals.csv')
```

```python
stop_data.head()
```

```python
plt.bar(stop_data['area_numbe'], stop_data['total_stops (updated)'])
plt.xlabel('Community Area')
plt.ylabel('Number of CTA Stops')
plt.title('Number of CTA Stops per Community Area (Buses and Trains)')
```

```python
plt.bar(stop_data['area_numbe'], stop_data['bus_count'])
plt.xlabel('Community Area')
plt.ylabel('Number of Bus Stops')
plt.title('Number of CTA Stops per Community Area (Buses)')
```

```python
plt.bar(stop_data['area_numbe'], stop_data['train_count'])
plt.xlabel('Community Area')
plt.ylabel('Number of Train Stops')
plt.title('Number of CTA Stops per Community Area (Trains)')
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4), dpi=300)


axes[0].bar(stop_data['area_numbe'], stop_data['total_stops (updated)'])
axes[0].set_xlabel('Community Area')
axes[0].set_ylabel('Number of CTA Stops')
axes[0].set_title('Buses and Trains')

axes[1].bar(stop_data['area_numbe'], stop_data['bus_count'])
axes[1].set_xlabel('Community Area')
axes[1].set_ylabel('Number of Bus Stops')
axes[1].set_title('Buses')

axes[2].bar(stop_data['area_numbe'], stop_data['train_count'])
axes[2].set_xlabel('Community Area')
axes[2].set_ylabel('Number of Train Stops')
axes[2].set_title('Trains')

fig.suptitle("CTA Stops per Community Area")

plt.savefig("Stop_Dist_Per_Community_Area.png", dpi = 300)
```

```python
plt.scatter(stop_data['train_count'], stop_data['bus_count'], alpha=0.6)
plt.xlabel('Number of Train Stops')
plt.ylabel('Number of Bus Stops')
plt.savefig('Train_Bus_Compare.png', dpi=300)
```
