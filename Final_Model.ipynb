### Significant Features from initial LR

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
lr_df = pd.read_csv('../Data/lr_sig_df.csv')
lr_df.drop(columns=['Unnamed: 0'], inplace = True)
lr_df
```

```python
dt_df = pd.read_csv('../Data/dt_imp.csv')
dt_df.drop(columns=['Unnamed: 0'], inplace= True)
dt_df.head()
```

```python
dt_df.rename(columns={'2020_HH_SIZE': 'hh_size'}, inplace=True)
dt_df.head()
```

```python
#Linear Regressio using ordinary least squares as error
model = smf.ols("""stop_dens_sqmi ~ Per_No_High_School_Diploma + per_hisp + per_black + MEDINC + Per_Unemployment""", 
                data = lr_df).fit()

#Results of model
model.summary()
```

```python
#Linear Regressio using ordinary least squares as error
model = smf.ols("""stop_dens_sqmi ~ pop_dens_sqmi + hh_size + Per_Unemployment + MEDINC + per_other""", 
                data = dt_df).fit()

#Results of model
model.summary()
```
